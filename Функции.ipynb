```python
# Вам нужно помочь секретарю автоматизировать работу. 
# Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.
# Исходные данные имеют следующую структуру:
#     перечень всех документов

documents = [
     {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
     {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
     {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
    ]

# перечень полок, на которых хранятся документы (если документ есть в documents, 
# то он обязательно должен быть и в directories)

directories = {
     '1': ['2207 876234', '11-2'],
     '2': ['10006'],
     '3': []
    }

# Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру

def holder():
    number = input('Введите номер документа: ')
    x = False
    for i in documents:
        if i['number'] == number:
            print('Владелец документа:', i['name'])
            x = True
            break
    if x == False:
        print('Документ не найден в базе')

#Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
def number():
    number = input('Введите номер документа: ')
    x = False
    len_ = len(list(directories.keys()))
    for i in range(1, len_ + 1):
        i = str(i)
        if number in directories[i]:
            print('Документ хранится на полке:', i)
            x = True
            break
    if x == False:
        print('Документ не найден в базе')

#Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам
def inf():
    len_ = len(list(directories.keys()))
    for d in documents:
        for i in range(1, len_ + 1):
            i = str(i)
            if d['number'] in directories[i]:
                print('№ ', d['number'], ', тип: ', d['type'], ', владелец: ', d['name'], ', полка хранения: ', i, sep='')
                break

#Пункт 4. Пользователь по команде "ads" может добавить новую полку
def ads():
    number = input('Введите номер полки: ')
    if number not in list(directories.keys()):
        directories[number] = []
        print('Полка добавлена. Текущий перечень полок: ', end='')
        print(*list(directories.keys()), sep=', ')
    else:
        print('Такая полка уже существует. Текущий перечень полок: ', end='')
        print(*list(directories.keys()), sep=', ')

#Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
def delete():
    number = input('Введите номер полки: ')
    if number in list(directories.keys()) and directories[number] == []:
        del directories[number]
        print('Полка удалена. Текущий перечень полок: ', end='')
        print(*list(directories.keys()), sep=', ')
    elif number in directories and directories[number] != []:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', end='')
        print(*list(directories.keys()), sep=', ')
    else:
        print('Такой полки не существует. Текущий перечень полок: ', end='')
        print(*list(directories.keys()), sep=', ')
        
print('Список доступных команд:')
print('по команде "p" можете узнать владельца документа по его номеру', 'по команде "s" можете по номеру документа узнать на какой полке он хранится', 'по команде "l" можете увидеть полную информацию по всем документам', 'по команде "ads" можете добавить новую полку', 'по команде "ds" можете удалить существующую полку из данных (только если она пустая)', 'по команде "q" - выход', sep='\n')

command = ''
while command != 'q':
    command = input('Введите команду: ')
    if command == 'p':
        holder()
    elif command == 's':
        number()
    elif command == 'l':
        inf()
    elif command == 'ads':
        ads()
    elif command == 'ds':
        delete()
```

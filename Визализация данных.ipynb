```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
happiness_score = pd.read_csv('2019.csv')
happiness_score.head()
```

```python
plt.barh(happiness_score['Score'], happiness_score['Perceptions of corruption'])
plt.xlabel('Perceptions of corruption')
plt.ylabel('Happiness Score')
plt.grid()
```

Можно сделать вывод, что даже в странах с высоким индексом счастья сильно развита коррупция.

И еще, что уровень счастья населения не зависит от уровня коррупции в стране.

```python
plt.plot(happiness_score['Score'], happiness_score['Healthy life expectancy'])
plt.xlabel('Happiness Score')
plt.ylabel('Healthy life expectancy')
plt.grid()
```

По графику видно, что в странах с высоким индексом счастья дольше ожидаемая продолжительность здоровой жизни населения.

```python
sns.boxplot(happiness_score['Freedom to make life choices'])
```

В большей части стран, включенных в отчет, есть свобода выбора.

```python
sns.pairplot(happiness_score[['Score', 'Social support', 'Generosity']])
```

Уровень счастья напрямую зависит от уровня социальной поддержки. А вот щедрость не зависит ни от уровня счастья, ни от социальной поддержки.
